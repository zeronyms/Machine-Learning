```python
import os
import shutil

# directory
source_dir = '/tmp/Application of Machine Learning in Detecting Iron Deficiency Anemia Using  Conjunctiva image Dataset from Ghana'
anemic_dir = os.path.join(source_dir, 'anemic')
non_anemic_dir = os.path.join(source_dir, 'non_anemic')

# buat directory
os.makedirs(anemic_dir, exist_ok=True)
os.makedirs(non_anemic_dir, exist_ok=True)

files = os.listdir(source_dir)
print(files)

# karena semua gambar menyatu dalam satu direktori, maka saya berusaha untuk memisahkan gambar sesuai dengan labelnya

for file in files:
    if 'Anemic' in file:
        shutil.move(os.path.join(source_dir, file), os.path.join(anemic_dir, file))
    elif 'Non-Anemic' in file or 'Non-anemic-Con' in file:
        shutil.move(os.path.join(source_dir, file), os.path.join(non_anemic_dir, file))

print("Pemisahan gambar selesai.")
```

```python
import tensorflow as tf
from keras.preprocessing.image import ImageDataGenerator
```

```python
datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=20,
    horizontal_flip=True,
    shear_range = 0.2,
    fill_mode = 'wrap',
    validation_split=0.4,
)
```

```python
train_generator = datagen.flow_from_directory(
    source_dir,
    target_size=(150, 150),
    class_mode = 'binary',
    shuffle = True,
    subset='training'
)
validation_generator = datagen.flow_from_directory(
    source_dir,
    target_size=(150, 150),
    class_mode = 'binary',
    shuffle = True,
    subset='validation'
)
```

```python
model = tf.keras.models.Sequential([
    tf.keras.layers.Conv2D(32, (3,3), activation='relu', input_shape=(150, 150, 3)),
    tf.keras.layers.MaxPooling2D(2, 2),
    tf.keras.layers.Conv2D(64, (3,3), activation='relu'),
    tf.keras.layers.MaxPooling2D(2,2),
    tf.keras.layers.Conv2D(128, (3,3), activation='relu'),
    tf.keras.layers.MaxPooling2D(2,2),
    tf.keras.layers.Conv2D(512, (3,3), activation='relu'),
    tf.keras.layers.MaxPooling2D(2,2),
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dropout(0.5),
    tf.keras.layers.Dense(512, activation='relu'),
    tf.keras.layers.Dropout(0.5),
    tf.keras.layers.Dense(1, activation='sigmoid')
])
```

```python
model.summary()
```

```python
model.compile(loss='binary_crossentropy',
              optimizer=tf.optimizers.Adam(),
              metrics=['accuracy'])
```

```python
model.fit(
      train_generator,
      steps_per_epoch=25,
      epochs=20,
      validation_data=validation_generator,
      validation_steps=5,
      verbose=2)
```
